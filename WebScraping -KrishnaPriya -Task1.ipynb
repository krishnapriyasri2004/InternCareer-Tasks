```python
data = []

for web1_gain in web_gain:
    prod_title = web1_gain.find('div', {'class': '_4rR01T'})
    prod1_title = prod_title.text.split('-')[0].strip()

    iphone_rating = web1_gain.find('div', {'class': '_3LWZlK'}).text

    try:
        Review = web1_gain.find('span', {'class': '_2_R_DZ'}).text
    except AttributeError:
        Review = None

    prod_mrp = web1_gain.find('div', {'class': '_30jeq3 _1_WHN1'}).text

    inform = web1_gain.find_all('li', {'class': 'rgWa7D'})
    Processor = inform[0].text
    ROM = inform[1].text
    Storage = inform[3].text

    Image_URL = web1_gain.img.get('src')

    data.append([prod1_title, iphone_rating, Review, prod_mrp, Processor, ROM, Storage,Image_URL])
```

```python
columns = ['prod_title', 'iphone_rating', 'Review', 'prod_mrp', 'Processor','ROM', 'Storage','Image']
df = pd.DataFrame(data, columns=columns)
print(df)
```
